```python
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud
import nltk
from nltk.corpus import stopwords
import string
```

```python
file_path = "IPL_Match_Highlights_Commentary.csv"  # Ensure this file is in the same folder
df = pd.read_csv(file_path)
```

```python
commentary_text = " ".join(df["Commentary"].dropna())
```

```python
stop_words.update(string.punctuation)
def clean_text(text):
    words = text.lower().split()  # Convert to lowercase and split words
    words = [word.strip(string.punctuation) for word in words if word not in stop_words]  # Remove stopwords & punctuation
    return " ".join(words)  # Rejoin words into a single string
cleaned_text = clean_text(commentary_text)
wordcloud = WordCloud(
    width=800, 
    height=400, 
    background_color="white"
).generate(cleaned_text)
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation="bilinear") 
plt.axis("off")  
plt.title("IPL Commentary Word Cloud", fontsize=14) 
plt.show() 
```
