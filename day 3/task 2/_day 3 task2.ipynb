```python
import pandas as pd
import matplotlib.pyplot as plt
file_path = "incomes.csv"  
df = pd.read_csv(file_path)
df_numeric = df.drop(columns=["country"]).astype(float)  
average_income = df_numeric.mean(axis=0)  
years = average_income.index.astype(int)
plt.figure(figsize=(10, 6))
plt.plot(average_income, years, marker='o', linestyle='-')
plt.xlabel("Average Income")
plt.ylabel("Year")
plt.title("Average Income Trend Over the Years")
plt.grid(True)
plt.show()
```
