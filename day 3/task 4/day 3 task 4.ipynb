```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_path = "cell_phones_total.csv"  
```

```python
df = pd.read_csv(file_path)
```

```python
df_melted = df.melt(id_vars=["country"], var_name="Year", value_name="Cell_Phones_Per_Person")
df_melted["Year"] = pd.to_numeric(df_melted["Year"], errors="coerce")
df_melted["Cell_Phones_Per_Person"] = pd.to_numeric(df_melted["Cell_Phones_Per_Person"], errors="coerce")
df_grouped = df_melted.groupby("Year", as_index=False)["Cell_Phones_Per_Person"].mean()
plt.figure(figsize=(10, 6))
plt.plot(df_grouped["Year"], df_grouped["Cell_Phones_Per_Person"], marker="o", linestyle="-", color="b")
plt.xlabel("Year")
plt.ylabel("Number of Cell Phones per Person")
plt.title("Rise in Cell Phone Usage Over the Years")
plt.grid(True)
plt.show()
```
